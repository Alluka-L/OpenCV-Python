```python
import cv2
from PIL import Image
```

```python
# read image
img = cv2.imread('./data/supra.jpeg')
img_1 = img.copy()

b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()

# BGR -> RGB
img[:, :, 0] = r
img[:, :, 1] = g
img[:, :, 2] = b
```

```python
Image.fromarray(img)
```

```python
img_rgb = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
Image.fromarray(img_rgb)
```

```python
# Save result
cv2.imwrite("./data/out.jpg", img)
```
