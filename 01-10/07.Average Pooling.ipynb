```python
import cv2
import numpy as np
from PIL import Image
```

```python
img = cv2.imread('./data/supra.jpeg')
out = img.copy()
```

```python
H, W, C = img.shape
G = 8
Nh = int(H / G)
Nw = int(W / G)

for y in range(Nh):
    for x in range(Nw):
        for c in range (C):
            out[G*y:G*(y+1), G*x:G*(x+1), c] = np.mean(out[G*y:G*(y+1), G*x:G*(x+1), c]).astype(np.int)
```

```python
Image.fromarray(out)
```
