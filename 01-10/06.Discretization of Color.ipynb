```python
import cv2
import numpy as np
from PIL import Image
```

```python
img = cv2.imread('./data/supra.jpeg')
```

```python
# Dicrease color
out = img.copy()

for i in range(4):
    ind = np.where(((64 * i - 1) <= out) & (out < (64 *(i + 1) - 1)))
    out[ind] = 32 * (2 * i + 1)
```

```python
Image.fromarray(out)
```
