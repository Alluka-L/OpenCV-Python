```python
import cv2
import numpy as np
from PIL import Image
```

```python
img = cv2.imread("../data/supra.jpeg")
H, W, C = img.shape

b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()
```

```python
# Gray scale
gray = 0.2126 * r + 0.7152 * g + 0.0722 * b
gray = gray.astype(np.uint8)
```

```python
Image.fromarray(gray)
```

```python
# Max-Min Filter
K_size = 3
```

```python
pad = K_size // 2
out = np.zeros((H + pad*2, W + pad*2), dtype=np.float)
out[pad:pad + H, pad:pad + W] = gray.copy()
tmp = out.copy()
```

```python
for y in range(H):
    for x in range(W):
        out[pad + y, pad + x] = np.max(tmp[y: y + K_size,x: x + K_size]) - np.min(tmp[y: y + K_size, x: x + K_size])
```

```python
out = out.astype(np.uint8)
```

```python
Image.fromarray(out)
```

```python
# Save result
cv2.imwrite("out13_MaxMinFilter.jpg", out)
```
