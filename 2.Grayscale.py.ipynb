```python
import cv2
import numpy as np
from PIL import Image
```

```python
# Read image
img = cv2.imread('./data/supra.jpeg')
img_origion = img.copy()

b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()
```

```python
# Grayscale
gray = cv2.cvtColor(img_origion, cv2.COLOR_BGR2GRAY)
Image.fromarray(gray)
```

```python
out = 0.2126 * r + 0.7152 * g + 0.0722 * b
```

```python
Image.fromarray(gray)
```

```python
cv2.imwrite('./data/out1.jpg", out)
```
