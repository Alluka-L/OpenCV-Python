```python
import cv2
import numpy as np
from PIL import Image
```

```python
img = cv2.imread('./data/supra.jpeg')
H, W, C = img.shape
```

```python
# Median Filter
K_size = 3
```

```python
# Zero padding
pad = K_size // 2
out = np.zeros((H + pad*2, W + pad*2, C), dtype=np.float)
out[pad:pad+H, pad:pad+W] = img.copy().astype(np.float)

for y in range(H):
    for x in range(W):
        for c in range(C):
            out[pad + y, pad + x, c] = np.median(out[y:y+K_size, x:x+K_size, c])
out = out[pad:pad+H, pad:pad+W].astype(np.uint8)
```

```python
Image.fromarray(out)
```
