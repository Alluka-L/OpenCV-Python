```python
import cv2
import numpy as np
from PIL import Image
```

```python
img = cv2.imread('./data/supra.jpeg').astype(np.float)
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()
```

```python
# Grayscale
out = 0.2126 * r + 0.7152 * g + 0.0722 * b
out = out.astype(np.uint8)
```

```python
# Binarization
th = 128
out[out < th] = 0
out[out >= th] = 255
```

```python
Image.fromarray(out)
```

```python
cv2.imwrite('./data/out_binarization.jpg', out)
```
